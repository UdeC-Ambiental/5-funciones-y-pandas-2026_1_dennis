## Desarrollo del reto
Aplica funciones y análisis según el objetivo. Usa `def` para definir tus funciones.

```python
import numpy as np

# 🔹 Función resumen
def resumen_termico(df):
    resumen = pd.DataFrame({
        "promedio": [df["temperatura"].mean(), df["humedad"].mean()],
        "maximo": [df["temperatura"].max(), df["humedad"].max()],
        "minimo": [df["temperatura"].min(), df["humedad"].min()]
    }, index=["temperatura", "humedad"])
    
    return resumen

print("Resumen:")
print(resumen_termico(df))


# 🔹 Función clasificar clima
def clasificar_clima(temp, hum):
    if temp < 20 and hum > 70:
        return "húmedo frío"
    elif temp < 20 and hum <= 70:
        return "seco frío"
    elif temp >= 20 and hum > 70:
        return "húmedo templado"
    else:
        return "seco templado"


# 🔹 Aplicar al DataFrame
df["clima"] = df.apply(lambda x: clasificar_clima(x["temperatura"], x["humedad"]), axis=1)

print("\nDataFrame con clima:")
print(df)


# 🔹 Tabla de frecuencias
tabla_clima = df["clima"].value_counts()

print("\nFrecuencia de tipos de clima:")
print(tabla_clima)
```

## Interpretación de resultados

## Interpretación de resultados

Se identificaron diferentes tipos de clima en los sitios analizados. 
Los climas húmedos están asociados a mayores niveles de humedad, mientras que los climas secos presentan menor humedad relativa.

Los resultados permiten observar que la temperatura y la humedad influyen directamente en la clasificación climática de cada sitio, lo cual puede afectar la biomasa y el comportamiento del ecosistema.
