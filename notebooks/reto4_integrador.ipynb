Objetivo: Integrar resultados y generar un reporte ambiental final.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/resultados_experimentos.csv')

df.head()
```

## Función de integración y reporte final

```python
def reporte_final(df):
    
    promedios = df[['temperatura', 'humedad', 'luz', 'biomasa']].mean()
    
    mayor = df.loc[df['biomasa'].idxmax()]
    
    menor = df.loc[df['biomasa'].idxmin()]
    
    correlaciones = df[['temperatura', 'humedad', 'luz', 'biomasa']].corr()['biomasa']
    
    variable = correlaciones.drop('biomasa').abs().idxmax()
    
    return promedios, mayor, menor, variable
```

```python
promedios, mayor, menor, variable = reporte_final(df)

print("PROMEDIOS")
print(promedios)

print("\nMAYOR BIOMASA")
print(mayor)

print("\nMENOR BIOMASA")
print(menor)

print("\nVARIABLE MÁS CORRELACIONADA")
print(variable)
```

## Gráfico de dispersión

```python
plt.scatter(df[variable], df['biomasa'])

plt.title(f'Biomasa vs {variable}')

plt.xlabel(variable)

plt.ylabel('Biomasa')

plt.show()
```

## Conclusión final

El análisis permitió identificar relaciones entre la biomasa y las variables ambientales. La variable más correlacionada con la biomasa puede influir directamente en la productividad de los ecosistemas analizados.
