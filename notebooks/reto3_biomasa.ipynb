Objetivo: Clasificar ecosistemas según su productividad y analizar variables ambientales asociadas.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/resultados_experimentos.csv')

df.head()
```

## Clasificación de biomasa

Se crea una función para clasificar la biomasa en niveles: alta, media y baja.

```python
def clasificar_biomasa(b):
    
    if b > 250:
        return 'alta'
    
    elif b >= 150:
        return 'media'
    
    else:
        return 'baja'
```

```python
df['nivel_biomasa'] = df['biomasa'].apply(clasificar_biomasa)

df.head()
```

## Promedios por nivel de biomasa

Se calculan los promedios de temperatura, humedad y luz para cada categoría de biomasa.

```python
promedios = df.groupby('nivel_biomasa')[['temperatura', 'humedad', 'luz']].mean()

promedios
```

```python
promedios.plot(kind='bar')

plt.title('Promedios ambientales por nivel de biomasa')

plt.xlabel('Nivel biomasa')

plt.ylabel('Promedio')

plt.show()
```

## Conclusión

Los ecosistemas con biomasa alta presentan diferencias en temperatura, humedad y luz respecto a los demás niveles. El análisis permitió identificar patrones ambientales relacionados con la productividad.
